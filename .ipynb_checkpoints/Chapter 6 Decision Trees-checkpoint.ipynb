```python
import sklearn as sk 
import numpy as np
import pandas as pd
import tensorflow as tf
import itertools

%matplotlib inline
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris = load_iris()
```

```python
X = iris.data[:,2:] #petal length and width
y = iris.target
```

```python
tree_clf = DecisionTreeClassifier(
    #max_depth=2
    min_samples_split=10
    )
tree_clf.fit(X,y)
```

```python
from sklearn.tree import export_graphviz
```

```python
export_graphviz(
    tree_clf,
    out_file = 'iris_tree.dot',
    feature_names= iris.feature_names[2:],
    class_names = iris.target_names,
    rounded = True,
    filled = True
    )
```

```python
def draw_tree_decision_boundard(clf, X, y, feature_names, class_names):

    import numpy as np
    import matplotlib.pyplot as plt
    import itertools
    
    
    # Parameters
    n_classes = len(np.unique(y))
    plot_colors = "ryb"
    plot_step = 0.02
    class_pairs = list(itertools.combinations(np.unique(y),2))
    plot_rows = np.ceil(len(class_pairs)/3)
    
    n_features = list(itertools.combinations(range(0,X.shape[1]),2))
    plot_cols = min(len(n_features),3)
    
    for pairidx, pair in enumerate(n_features):
        # We only take the two corresponding features
        P = X[:, list(pair)]
        
        
        # Plot the decision boundary
        plt.subplot(plot_rows, 1, pairidx + 1, aspect= 'equal')

        x_min, x_max = P[:, 0].min() - 1, P[:, 0].max() + 1
        y_min, y_max = P[:, 1].min() - 1, P[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                             np.arange(y_min, y_max, plot_step))
        plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        cs = plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu)

        plt.xlabel(feature_names[pair[0]])
        plt.ylabel(feature_names[pair[1]])

        # Plot the training points
        for i, color in zip(range(n_classes), plot_colors):
            idx = np.where(y == i)
            plt.scatter(P[idx, 0], P[idx, 1], c=color, label=class_names[i],
                        cmap=plt.cm.RdYlBu, edgecolor='black', s=15)

    plt.suptitle("Decision surface of a decision tree using paired features")
    plt.legend(loc='best', borderpad=0, handletextpad=0)
    plt.axis("tight")
    plt.show()
    return plt
```

```python
feature_names= iris.feature_names[2:]
class_names = iris.target_names
draw_tree_decision_boundard(tree_clf,X,y, feature_names, class_names)
```
