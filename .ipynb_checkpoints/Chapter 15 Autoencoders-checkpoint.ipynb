```python
#Build a linear activiation function neural network autoencoder
#that is the same as a PCA analysis
```

```python
from tensorflow.contrib.layers import fully_connected

n_inputs = 3
n_hidden = 2
n_outputs = n_inputs

learning_rate = 0.01


#define the variables
X = tf.placeholder(tf.float32, shape=[None,n_inputs])

#define the neural network
hidden_layers = fully_connected(
    inputs = X,
    num_outputs = n_hidden,
    activation_fn = None
    )

output_layers = fully_connected(
    inputs = hidden_layers,
    num_outputs = n_outputs,
    activation_fn = None
    )

#define the loss model
loss = tf.losses.mean_squared_error(X,output_layers)

#define the optimizer & train step
optimizer = tf.train.AdamOptimizer(learning_rate)
train_step = optimizer.minimize(loss)

#train the model
n_iterations = 10000

codings = hidden_layers #define intermediate variable outside
                        #the model so can evaluate it and recover
                        #the encoded data

with tf.Session() as sess:
    tf.global_variables_initializer().run()
    for i in range(n_iterations):
        train_step.run(feed_dict = {X: X_train})
        if i % 1000 == 0:
            loss_val = loss.eval(feed_dict={X: X_train})
            print('Iteration {} Loss:{:.4f}'.format(i,loss_val))
    codings_val = codings.eval(feed_dict={X: X_train})
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax = ax.scatter(codings_val[:,0],codings_val[:,1],c=X_train[:,2], cmap='jet')
plt.suptitle('Reduced 2D Data Swiss Roll')
plt.colorbar(ax)
plt.show()
```

```python
def load_mnist():
    """Function to download mnst data from public source.
        Args:
            None

        Returns:
            train_set --tuple of data and target

            valid_set --tuple of data and target

            test_set --tuple of data and target
    """
    import urllib.request, pickle, gzip
    r = urllib.request.urlretrieve('http://deeplearning.net/data/mnist/mnist.pkl.gz','mnist.pkl.gz')
    filename = r[0]
    print(filename)
    with gzip.open(filename, 'rb') as f:
        train_set, valid_set, test_set = pickle.load(f,encoding='bytes')
        f.close()
    return train_set, valid_set, test_set

train_set, valid_set, test_set = load_mnist()

X_train = train_set[0]
y_train = train_set[1]

# X_valid = valid_set[0]
# y_valid = valid_set[1]

# X_test = test_set[0]
# y_test = test_set[1]
```

```python
def plot_digits(data, images_per_row=None):
    from matplotlib.cm import binary
    from numpy import sqrt, floor
    images_per_row = images_per_row or int(np.ceil(np.sqrt(len(data))))
    rows = int(np.ceil(len(data)/images_per_row))
    fig = plt.figure()
    i = 0
    j = 0
    k=0
    dim = int(floor(sqrt(data.shape[1])))
    dim2 = max([n for n in range(1,dim+1) if data.shape[1] % n == 0])
    dim1 = int(data.shape[1]/dim2)
    print('{} pixels displayed as {}x{} grid'.format(data.shape[1],dim1,dim2))
    #dim2 = int(floor(sqrt(data.shape[1])))
    for d in data:
        i = int(np.floor(k/images_per_row))
        j = k % images_per_row
        k = k+1
        #d = d.reshape(15, 10)
        d = d.reshape(dim1, dim2)
        #print(i,j,images_per_row,rows)
        ax = plt.subplot2grid((rows,images_per_row),(i,j))
        ax.imshow(d, cmap = binary,
        interpolation="nearest")
        plt.axis("off")
       
```

```python
#build a concise model for the MNIST learning
import time
from tensorflow.contrib.layers import fully_connected


def random_batch(X_train, y_train, batch_size):
    rnd_indices = np.random.randint(0, len(X_train), batch_size)
    X_batch = X_train[rnd_indices]
    y_batch = y_train[rnd_indices]
    return X_batch, y_batch


#define the parameters learning
n_inputs = X_train.shape[1]
n_hidden1 = 300
n_hidden2 = 150
n_hidden3 = n_hidden1
n_outputs = n_inputs

learning_rate = 0.01
l2_reg = 0.001

#define the variables
X = tf.placeholder(tf.float32, shape=[None,n_inputs])

#define the neural network
with tf.contrib.framework.arg_scope(
    [fully_connected],
    activation_fn = tf.nn.elu,
    weights_initializer = tf.contrib.layers.variance_scaling_initializer(),
    weights_regularizer = tf.contrib.layers.l2_regularizer(l2_reg)
    ):

    hidden_layer1 = fully_connected(
        inputs = X,
        num_outputs = n_hidden1
        )

    hidden_layer2 = fully_connected(
        inputs = hidden_layer1,
        num_outputs = n_hidden2
        )

    hidden_layer3 = fully_connected(
        inputs = hidden_layer2,
        num_outputs = n_hidden3
        )

    output_layers = fully_connected(
        inputs = hidden_layer3,
        num_outputs = n_outputs,
        )

#define variable handles for easy evaluation post-training    
codings = hidden_layer2 
encodings = output_layers

#define the loss model
reconstruction_loss = tf.losses.mean_squared_error(X,output_layers)

reg_losses = tf.get_collection(tf.GraphKeys.REGULARIZATION_LOSSES)
loss = tf.add_n([reconstruction_loss] + reg_losses)

#define the optimizer & train step
optimizer = tf.train.AdamOptimizer(learning_rate)
train_step = optimizer.minimize(loss)

#batch train the model
n_epoch = 10
batch_size = 150
n_batches = int(np.ceil(X_train.shape[0] / 150))

runtime_stats =  []

with tf.Session() as sess:
    tf.global_variables_initializer().run()
    start = time.time()
    for epoch in range(n_epoch):
        for iteration in range(n_batches):
            X_batch, y_batch = random_batch(X_train, y_train, batch_size)
            train_step.run(feed_dict = {X: X_batch})
            loss_val = loss.eval(feed_dict={X: X_batch})
            timer = time.time()-start
            runtime_stats.append([epoch, iteration,loss_val, timer])
            
            if iteration % 100 == 0:
                print('Epoch:{} Iteration:{} Loss:{:.4f} Runtime:{:.2f} secs'.format(*runtime_stats[-1]))
    codings_val = codings.eval(feed_dict={X: X_train})
    encodings_val = encodings.eval(feed_dict={X: X_train})
    runtime_stats = np.array(runtime_stats)
    print('END Epoch:{} Iteration:{} Loss:{:.4f} Runtime:{:.2f} secs'.format(*runtime_stats[-1]))
```
